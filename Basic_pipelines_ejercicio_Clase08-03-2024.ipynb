```python
# Pipeline
from sklearn.pipeline import Pipeline

# Importamos un clasificador: Support Vector Classifier
from sklearn.svm import SVC

# Scaler estandar
from sklearn.preprocessing import StandardScaler

# Splitter
from sklearn.model_selection import train_test_split

from sklearn.decomposition import PCA
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

import pandas as pd
```

```python
# Genera un problema de clasificación randmom de n-classes.

from sklearn.datasets import make_classification


data = pd.read_excel(r"D:\Download\Base.XLSX", sheet_name='Sheet1')

data
```

```python
data['Fecha fabricacion'] = pd.to_datetime(data['Fecha fabricacion'])
```

```python
mask = (data['CENTRO']!=1000)
mask2 = ((data['Fecha fabricacion'].dt.month==1) | (data['Fecha fabricacion'].dt.month==2))
mask3 = (data['%.vida.Util.Meses']>90)
```

```python
data['target']= 0
```

```python
data['mes_fabricacion'] = pd.to_datetime(data['Fecha fabricacion']).dt.month
data['dia_fabricacion'] = pd.to_datetime(data['Fecha fabricacion']).dt.day

data['mes_consumo'] = pd.to_datetime(data['Fecha,pref.consumo']).dt.month
data['dia_consumo'] = pd.to_datetime(data['Fecha,pref.consumo']).dt.day

data.drop(columns=['Fecha,pref.consumo','Fecha fabricacion'],inplace=True)
```

```python
X = data[['Material', 'Grupo.Articulo', 'Grupo', 'DESCRIPCION', 'UNIDAD',
       'CENTRO', 'ALMACEN', 'LOTE',  '%.vida.Util.Meses',
       'Vida.Util.Meses',  'LIBRE UTIL.', 'CALIDAD',
       'BLOQUEADO', 'STOCK TOTAL', 'Peso neto', 'Costo posicion', 'Prom.Venta',
       'Averias', 'Vencidos', 'Por Vencer', 'Ajustes', 'Ajustes Transp.',
       'Tmpo.mín.dur(Meses)', 'Tmpo.mín.dur(Dias)', 'Stock No Libre',
       'Reproc.Producion', 'Rev.cliente','mes_fabricacion', 
       'dia_fabricacion',
       'mes_consumo', 'dia_consumo']]
y = data[['target']]
```

```python
from sklearn.compose import ColumnTransformer
categorical_features = ['Grupo', 'DESCRIPCION','UNIDAD','LOTE','ALMACEN']  #columnas categóricas

#transformador de columnas, aplicando OneHotEncoder solo a las columnas categóricas
preprocessor = ColumnTransformer(
    transformers=[
        ('one_hot_encoder', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough'  # Esto permite que el resto de las columnas pasen sin cambios
)
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
# Dividimos en test y train
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# Crea el pipeline
pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('scaler', StandardScaler(with_mean=False)),        
    #('pca', PCA(n_components=2)),  
    ('svd', TruncatedSVD(n_components=100)),      
    ('classifier', GradientBoostingClassifier())  
])

# El pipeline puede usarse como un estimador
# y evita data leaking del test set en el train set

# Se aplican los pasos en el dataset de entrenamiento
pipe.fit(X_train, y_train)

# se aplican los pasos en el dataset de prueba
pipe.score(X_test, y_test)
```
